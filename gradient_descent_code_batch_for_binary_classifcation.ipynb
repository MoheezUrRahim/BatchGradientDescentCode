**Plain code of Gradient Descent Algortim (Batch)**

```python
import numpy as np
```

```python
def sigmoid(z):
  return 1/(1-np.exp(-z))
```

```python
def log_loss(y,y_pred,epsilon=1e-15):
  y_pred=np.clip(y_pred,epsilon,1-epsilon)
  return -np.mean(y*np.log(y_pred)+(1-y)*np.log(1-y_pred))
```

```python
import numpy as np
def gradient_descent(x_1,x_2,y,epoch,learning_rate,threshold=1e-6):
  w_1=w_2=1
  b=0
  n=len(x_1)
  for i in range(epoch):
    y_pred=np.array(sigmoid(w_1*x_1+w_2*x_2+b))
    loss=log_loss(y,y_pred)
    w_1=w_1-learning_rate*np.dot(np.transpose(x_1),(y_pred-y))*(1/n)
    w_2=w_2-learning_rate*np.dot(np.transpose(x_2),(y_pred-y))*(1/n)
    b=b-learning_rate*np.mean(y_pred-y)
    print (f'Epoch:{i}, w1:{w_1}, w2:{w_2}, bias:{b}, loss:{loss}')
    if loss<=threshold:
      break
  return w_1,w_2,b,loss
```

```python

#example code:
x_1=np.array([1,2,3,4,5])
x_2=np.array([1,3,6,10,9])
y=np.array([0,1,0,1,1])

a,b,c,d=gradient_descent(x_1,x_2,y,100,0.9)
```
